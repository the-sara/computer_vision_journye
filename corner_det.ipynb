```python
import cv2
import matplotlib.pyplot as plt 
img=cv2.imread("daizy2.jpg")
plt.imshow(img)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
corners1 = cv2.cornerHarris(img_gray, blockSize=3, ksize=3, k=0.04)
plt.imshow(corners1)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
corners2 = cv2.cornerHarris(img_gray, blockSize=11, ksize=11, k=0.04)
plt.imshow(corners2)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
corners3 = cv2.cornerHarris(img_gray, blockSize=3, ksize=3, k=2)
plt.imshow(corners3)
```
