```python
import cv2
import matplotlib.pyplot as plt 
```

```python
eye_cascade  = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
img=cv2.imread("tesla_pic.jpg")
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
rectangle=eye_cascade.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=10)
for (x,y,w,h) in rectangle :
    cv2.rectangle(img,(x,y),(x+w,y+h),(200,10,100),20)
plt.imshow(img)
```
