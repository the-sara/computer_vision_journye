```python
import cv2
import matplotlib.pyplot as plt 
img=cv2.imread("daizy2.jpg")
plt.imshow(img)
```

```python
# instantiation :
sift=cv2.SIFT_create()
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
kp,des=sift.detectAndCompute(img_gray,None)
```

```python
img_kp=cv2.drawKeypoints(img,kp,None)
plt.imshow(img_kp)
```
