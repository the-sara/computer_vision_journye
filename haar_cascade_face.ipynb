```python
import cv2
import matplotlib.pyplot as plt 
```

```python
#instatiate the pretrained model from CascadeClassifier class
face_detction_model=cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
#getting the img ready:
img=cv2.imread("tesla_pic.jpg")
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
#the pretrained models has a methode 
#detectMultiScale:returns a list of triangles(x,y,w,h)
rectangle=face_detction_model.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=5)
```

```python
# drawing the rectangle on the detected face :
for (x,y,w,h) in rectangle :
    cv2.rectangle(img,(x,y),(x+w,y+h),(200,10,100),20)
```
