```python
import cv2
import matplotlib.pyplot as plt 
img=cv2.imread("daizy2.jpg")
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
#instatiation:
fast=cv2.FastFeatureDetector_create()
```

```python
kp=fast.detect(img_gray,None)
```

```python
img_kp=cv2.drawKeypoints(img,kp,None)
plt.imshow(img_kp)
```
