```python
import cv2
import matplotlib.pyplot as plt 
```

```python
# getting the img:
img=cv2.imread("daizy2.jpg")
plt.imshow(img)
```

```python
# converting to gray scale
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
```

```python
#converting to hsv:
img_hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```
